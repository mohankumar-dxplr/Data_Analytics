```python
import plotly.express as px

division_units = (
    df.groupby("Division", as_index=False)["Units"]
      .sum()
)

fig = px.bar(
    division_units,
    x="Division",
    y="Units",
    title="Total Units by Division"
)

fig.update_layout(
    width=1000,
    height=600
)

fig.show()
```
